## Loading CSV & Exploratory Data Analysis



```python
movie=pd.read_csv("top10K-TMDB-movies.csv")
```

```python
movie.head()
```

```python
movie.info()
```

## Feature Engineering

```python
movie= movie.drop(columns=["original_language","popularity","release_date","vote_average","vote_count"])
```

```python
movie["tags"]=movie["overview"]+movie["genre"]
```

```python
movie.drop(columns=["overview","genre"],inplace=True)
```

```python
movie.info()
```

```python
movie['tags']=movie["tags"].fillna("None")
```

## Converting text into vector

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector=cv.fit_transform(movie['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
distance=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda vector:vector[1])
for i in distance[1:10]:
  print(movie.iloc[i[0]].title)
```

## Making Movie recommender funtion

```python
def recommender(movies):
  index=movie[movie['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  movies_list=[]
  for i in distance[1:10]:
    print(movie.iloc[i[0]].title)
```

```python
recommender("The Dark Knight")
```
